# Na trajectories

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

#Set number of Na atoms, timesteps and extract lines with Na

step_count = []
na = []

with open('long.lammpstrj') as file:
    content = file.readlines()
for line in content:
    if 'TIMESTEP' in line:
        step_count.append(line)

with open('long.lammpstrj') as file:
    content = file.readlines()
for line in content:
    if 'Na ' in line:
        na.append(line)

steps = len(step_count)
na_atoms = int(len(na) / len(step_count))

#Sort out coordinates by dimension

na_coords = [[] for x in range(0, na_atoms)]
na_coords_split = [[] for x in range(0, na_atoms)]
na_coords_x = [[] for x in range(0, na_atoms)]
na_coords_y = [[] for x in range(0, na_atoms)]
na_coords_z = [[] for x in range(0, na_atoms)]
    
for n in range(0, na_atoms):
    for i in range (n, steps * na_atoms, na_atoms):
        na_coords[n].append(na[i])
    for i in na_coords[n]:
        na_coords_split[n].append(i.split())
    for i in range(0, steps):
        na_coords_x[n].append(float(na_coords_split[n][i][1]))
    for i in range(0, steps):
        na_coords_y[n].append(float(na_coords_split[n][i][2]))
    for i in range(0, steps):
        na_coords_z[n].append(float(na_coords_split[n][i][3]))
        
#Create figure

fig = plt.figure(figsize=(20,20))
(ax, ax2, ax3) = fig.subplots(1, 3)

ax.set_title('xy projection', size=20)
ax.set_xlim(-1, 14.504)
ax.set_ylim(-1, 14.504)
ax.set_aspect('equal', adjustable='box')

ax2.set_title('xz projection', size=20)
ax2.set_xlim(-1, 14.504)
ax2.set_ylim(-1, 14.504)
ax2.set_aspect('equal', adjustable='box')

ax3.set_title('yz projection', size=20)
ax3.set_xlim(-1, 14.504)
ax3.set_ylim(-1, 14.504)
ax3.set_aspect('equal', adjustable='box')

#Plot paths

for n in range(0, na_atoms):
    ax.plot(na_coords_x[n], na_coords_y[n], color='orange', zorder=-1)

for n in range(0, na_atoms):
    ax2.plot(na_coords_x[n], na_coords_z[n], color='orange', zorder=-1)

for n in range(0, na_atoms):
    ax3.plot(na_coords_y[n], na_coords_z[n], color='orange', zorder=-1)

#Add oxide ions

o_x = [2.250500045010, 6.751500000000, 11.252499954990]
o_y = [2.250500045010, 6.751500000000, 11.252499954990]
o_coords = []
o_coords2 = []
o_coords3 = []

for i in o_x:
    for j in o_y:
        o_coords.append(patches.Circle((i, j), 0.5, facecolor='red', edgecolor='black'))
        o_coords2.append(patches.Circle((i, j), 0.5, facecolor='red', edgecolor='black'))
        o_coords3.append(patches.Circle((i, j), 0.5, facecolor='red', edgecolor='black'))
    
for p, q, r in zip(o_coords, o_coords2, o_coords3):
    ax.add_patch(p)
    ax2.add_patch(q)
    ax3.add_patch(r)
    
#Add chloride ions 

cl_x = [0, 4.500999954990, 9.002000045010, 13.504]
cl_y = [0, 4.500999954990, 9.002000045010, 13.504]
cl_coords = []
cl_coords2 = []
cl_coords3 = []

for i in cl_x:
    for j in cl_y:
        cl_coords.append(patches.Circle((i, j), 0.5, facecolor='green', edgecolor='black'))
        cl_coords2.append(patches.Circle((i, j), 0.5, facecolor='green', edgecolor='black'))
        cl_coords3.append(patches.Circle((i, j), 0.5, facecolor='green', edgecolor='black'))
    
for p, q, r in zip(cl_coords, cl_coords2, cl_coords3):
    ax.add_patch(p)
    ax2.add_patch(q)
    ax3.add_patch(r)
```
